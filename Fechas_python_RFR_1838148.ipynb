## Fechas en numpy

```python
import numpy as np
```

```python
## por default reconoce dias en datetime64
np.array(['1997-12-12','1998-10-01','2022-03-29'], dtype='datetime64')
```

```python
## podemos especificar que reconozca horas
np.array(['1997-12-12','1998-10-01','2022-03-29'], dtype='datetime64[h]')
```

```python
## especificar que reconozca el año
np.array(['1997-12-12','1998-10-01','2022-03-29'], dtype='datetime64[Y]')
```

```python
## vamor a pedir a numpy
np.arange('2023-01-01','2023-06-30',30,dtype='datetime64[D]')
```

```python
np.arange('1997-12-12','2023-03-30',1,dtype='datetime64[Y]')
```

```python
# importar la libreria yahoo finance
import yfinance as yf 
```

```python
tipo_cambio= yf.download('MXN=X',
                         start='2015-01-01',
                         end='2022-12-31')
```

```python
tipo_cambio= tipo_cambio[['Adj Close']]
```

```python
tipo_cambio
```

```python
## Graficar
tipo_cambio.plot();
```

```python
tipo_cambio.index
```

```python
# al tipo de cambio diario le vamos a cambiar la frecuencia
# tipo de cambio anual, promedio 
tipo_cambio_anual = tipo_cambio.resample(rule='AS').mean()
tipo_cambio_anual
```

```python
## tipo de cambio mensual
tipo_cambio_mensual = tipo_cambio.resample(rule='MS').mean()
tipo_cambio_mensual
```

```python
# dias habiles = bussiness day 'B'
tipo_cambio = tipo_cambio.asfreq('B')
```

```python
tipo_cambio.index
```

```python
# valor maximo del tipo de cambio de cada mes
tipo_cambio.resample(rule='MS').max().plot();
```
