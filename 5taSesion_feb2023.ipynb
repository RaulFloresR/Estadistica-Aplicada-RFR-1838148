```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5]
y=[1,8,27,64,125]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
```

```python
plt.plot(x,y,color='red')
```

```python
mun_nl= pd.DataFrame({'municipio':['Monterrey','Guadalupe','Apodaca','San Nicolás'],'nacimientos':[500,630,768,400]})
```

```python
mun_nl
```

```python
plt.bar(mun_nl['municipio'],mun_nl['nacimientos'])
```

```python
plt.bar(mun_nl['municipio'],mun_nl['nacimientos'],color=['blue','green','red','yellow'])
plt.title('Nacimientos por municipio')
plt.xlabel('Municipios de NL')
plt.ylabel('Cantidad de nacimientos')
plt.yticks([0,200,400,600,800])
plt.show()
```

```python
#lo mismo pero ordenado
mun_nl = mun_nl.sort_values('nacimientos',ascending=False)

plt.bar(mun_nl['municipio'],mun_nl['nacimientos'],color=['blue','green','red','yellow'])
plt.title('Nacimientos por municipio')
plt.xlabel('Municipios de NL')
plt.ylabel('Cantidad de nacimientos')
plt.yticks([0,200,400,600,800])
plt.show()
```

```python
fig, (panel1,panel2)=plt.subplots(2,1)
fig.tight_layout()

#panel 1
x1=x
y1=y
panel1.plot(x,y,label='nombre de linea')
panel1.title.set_text('Grafica 1')

#panel 2 
x2=mun_nl['municipio']
y2=mun_nl['nacimientos']
panel2.bar(x2,y2)
panel2.title.set_text('Grafica 2')

plt.show()
```

```python
fig, (panel1,panel2)=plt.subplots(1,2)
fig.tight_layout()

#panel 1
x1=x
y1=y
panel1.plot(x,y,label='nombre de linea')
panel1.title.set_text('Grafica 1')

#panel 2 
x2=mun_nl['municipio']
y2=mun_nl['nacimientos']
panel2.bar(x2,y2)
panel2.title.set_text('Grafica 2')

plt.show()
```

```python
mun_nl = mun_nl.sort_values('nacimientos',ascending=True)

plt.barh(mun_nl['municipio'],mun_nl['nacimientos'],color=['blue','green','red','yellow'])
plt.title('Nacimientos por municipio')
plt.xlabel('Municipios de NL')
plt.ylabel('Cantidad de nacimientos')

plt.show()
```
