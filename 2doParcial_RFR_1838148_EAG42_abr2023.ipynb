1. El banco sospecha que entre mas pasa el tiempo más díficil es que puedan pagar los clientes que se les concedió un préstamo. 

a) Calcula el promedio del monto del préstamo de los que pagaron y el promedio del monto de los que no han pagado.

b) Ajusta un modelo weibull y un modelo log-normal y graficas las funciones de riesgo respectivas. 


```python
# a)
credit.groupby('full_repaid').agg('mean')
```

```python
# b)
from lifelines import WeibullFitter

Weib= WeibullFitter().fit(credit['duration'],credit['full_repaid'])
Weib.plot_hazard();
```

```python
from lifelines import LogNormalFitter
logn=LogNormalFitter().fit(credit['duration'],credit['full_repaid'])
logn.plot_hazard();
```

d) comenta que significa la grafica de la función de riesgo de la log-normal respecto a si los clientes tardan en pagar

La gráfica log-normal dice que el mayor riesgo de que una persona no complete su pago se va a presentar entre los meses 20 y 30, sin embargo bajará un poco después.

c)

```python
#d)
from lifelines.statistics import logrank_test
```

h0: $S_1(t)=S_2(t)$

Ha: $S_1(t)\neq S_2(t)$

```python
lrt= logrank_test(mucho['duration'],mucho['full_repaid'],
                  poco['duration'],poco['full_repaid'])
lrt.print_summary()
```

Como $p<\alpha=0.05$ podemos decir con un 95% de confianza que rechazamos H0, lo que significa que las funciones de supervivencia de los que deben mucho y poco son distintas

6. Elige seis variables para ajustar un modelo de cox de riesgos porporcionales. Antes de correr el modelo explica en qué sentido y por qué afectan el riesgo de pagar el préstamo cada una de las variables elegidas


Las variables que se van a utilizar para el modelo de Cox son:

1. age: debido a que las personas jovenes pueden estar mas incentivadas a mantener un buen historial crediticio para el futuro o que las personas mayores tengan un trabajo mas estable que les permita pagar 

2.  people_liable: debido a que tal vez una persona que tenga mas gente a su cargo tiene mayor dificultad para realizar sus pagos por los gastos extra que esto representa

3. number_of_credits: ya que si el banco le autoriza mas creditos significa que es una persona que si puede pagarlos 

4. telephone: porque al ser practicamente una necesidad, quien no tenga telefono significa que no tendra para pagar un prestamo

5. foreign_worker: porque una persona trabajando en el extranjero se sentira con menos obligacion de pagar un credito

6. installment_rate: porque a una persona con un interes mas alto le sera mas dificil pagar que a una con un interes bajo

```python
from lifelines import CoxPHFitter
```

```python
credit_c= pd.DataFrame(credit)
credit_c= credit_c[['duration','full_repaid','age','people_liable','number_of_credits','telephone','foreign_worker','installment_rate']]
credit_c.head()
```

```python
cph=CoxPHFitter().fit(credit_c,'duration','full_repaid')
cph.print_summary()
```

7. Interpreta la regresión de Cox

Al verificar el valor p, notamos que tanto people_liable como number_of_credits no son significativas por su alto valor en esta columna.

La edad aumenta la probabilidad de que se complete el pago, con un aumento en promedio de 1% por cada año cumplido.

Una persona que trabaja en el extranjero tiene un 68% mas de probabilidad de no pagar que una que no.

Por último, en promedio se tiene una probabilidad 10% mas alta de que una persona no pague por cada porcentaje de interes que se les cobre.
