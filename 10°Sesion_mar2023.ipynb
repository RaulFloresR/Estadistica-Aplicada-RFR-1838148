 signo "coef" nos va a indicar si el riesgo aumenta o disminuye.
             Si coef < 0  ->  1-exp(coef) = %
             Si coef > 0  ->  exp(coef)-1 = %
             
Recibir ayudar (fin) disminuye el riesgo de volver a ser arrestado.

La edad dismuye el riesgo de volver a ser arrestado.
1-0.94 = 6%
Por cada ano cumplido en promedio, el riesgo de volver a ser arrestado disminuye 6%.

Haber sido arrestado en el pasado aumenta el riesgo de volver a ser arrestado.
