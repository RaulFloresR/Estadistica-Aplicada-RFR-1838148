```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
%matplotlib inline
#pip instal statsmodels si no esta instalada
```

```python
datos = pd.read_excel('C:/Users/raulf/Desktop/8° Semestre/Estadistica aplicada/modelolineal.xlsx')
```

```python
datos.shape
```

```python
datos.head()
```

```python
datos = datos.set_index('observación')
```

```python
datos.head()
```

```python
datos.describe()
```

```python
datos.plot.scatter('X', 'Y');
plt.title('Gráfica de deispersión entre X y Y')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.xlim(0, 30) #limitar el eje X
plt.ylim(0, 30)

plt.show()
```

```python
reg = smf.ols(formula='Y~X', data = datos) #regresion por minimos cuadrados ordinarios
results = reg.fit()

print(results.summary())
```

```python
datos['X_sq'] = datos['X']**2 
```

```python
datos.head()
```

```python
reg2 = smf.ols(formula='Y~X+X_sq', data=datos)
results2 = reg2.fit()

print(results2.summary())
```
